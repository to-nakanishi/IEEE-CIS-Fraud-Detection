特徴量変換の方向性について  
以下の実施時に、有効となる特徴量があれば追加または変換を行う。

1.   オブジェクト型の確認  
LightGBM試走でエラーとならなかったため、存在しない可能性が高いが、リスク排除のため確認を行う。
1.   ストリング型  
現時点では、LightGBMではNanかそれ以外かで学習しているため、カテゴリー型に変換を実施。特徴量は以下のカテゴリー型でまとめて検討を行う。


1.   カテゴリー型  
モデルで学習できるため型変換は不要であるが、有効な特徴量の追加ができないか検討を行う。


2.   int,float型  
ばらつきが大きいものに関しては通常、対数変換することが多いが、不均衡モデルに関してはばらつき自体が有力な情報である可能性もあるため、変換ではなく追加するほうがリスクは少ない。

```python
#2-2　特徴量変換　カテゴリー型のカラム確認

# 1. カテゴリ型のカラム名を取得
cat_cols = train.select_dtypes(include=['category']).columns

# 2. カラム名の一覧と、それぞれの「ユニークな値の数」を表示
print(f"📊 カテゴリ型カラム数: {len(cat_cols)}")
print("-" * 30)

for col in cat_cols:
    unique_count = train[col].nunique()
    # 最初の5つだけ具体例を表示して雰囲気をつかむ
    example_values = train[col].dropna().unique()[:5].tolist()
    print(f"{col:20} | ユニーク数: {unique_count:5} | 例: {example_values}")
```

```python
#2-2　特徴量変換  ユニーク値の確認(P_emaildomain,R_emaildomain)

def analyze_fraud_by_category(df, col, min_count=10, top_n=25):
    """
    指定したカラムの不正率を分析し、上位・下位を表示する
    """
    print(f"\n🔥 --- {col} 分析結果 (Top {top_n} & Tail {top_n}) ---")

    # 1. 各カテゴリの集計
    # インデックスを文字列として扱う（NaN対策）
    summary = df.groupby(df[col].astype(str), dropna=False)['isFraud'].agg(['count', 'mean'])
    summary['割合(%)'] = (summary['count'] / len(df)) * 100
    summary['不正率(%)'] = summary['mean'] * 100

    # 2. 最小件数以上のデータに絞ってソート
    valid_summary = summary[summary['count'] >= min_count].sort_values('不正率(%)', ascending=False)

    # 3. 上位と下位を抽出
    top_df = valid_summary.head(top_n)
    tail_df = valid_summary.tail(top_n)

    # 4. 表示
    print(f"【不正率が高い(Top{top_n}) / 低い(Tail{top_n}) / 最小件数: {min_count}】")
    # 分離ライン用のダミー行を作成
    separator = pd.DataFrame(index=['--------------------'], columns=top_df.columns)
    display_df = pd.concat([top_df, separator, tail_df])

    print(display_df[['count', '割合(%)', '不正率(%)']])

    # 5. NaNの状況（インデックスが 'nan' になっているものを探す）
    print(f"\n【📍 欠損値(NaN) の状況】")
    nan_info = summary[summary.index.str.lower().isin(['nan', 'none', 'null'])]
    if not nan_info.empty:
        print(nan_info[['count', '割合(%)', '不正率(%)']])
    else:
        print("NaNデータは見当たりません。")

    print("-" * 60)
    print(f"\n💡 (参考) 全体の平均不正率: {train['isFraud'].mean()*100:.3f}%")

for col in ['P_emaildomain', 'R_emaildomain']:
  analyze_fraud_by_category(train,col)
```

OSやブラウザと比較して、名前ソート順で見たときに不正率0のものが多く、不正で使用される解像度への偏りが見られた。可視化を行うため縦横を乗算し、対数スケールでヒストグラムを行う。


```python
#2-3　特徴量変換  ユニーク値の確認(id_33)　解像度変換及び対数変換ヒストグラム(不正値、正常値比較)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 解像度(id_33)を「面積」に変換する関数（変更なし）
def get_resolution_area(res_str):
    if pd.isna(res_str) or res_str == 'nan':
        return np.nan
    try:
        w, h = map(int, res_str.split('x'))
        return w * h
    except:
        return np.nan

# 2. 面積カラム作成と対数変換
train['Res_Area'] = train['id_33'].apply(get_resolution_area)
train['Res_Area_log'] = np.log1p(train['Res_Area'])

# 3. ヒストグラムの描画（視認性向上版）
plt.figure(figsize=(14, 7))

# 正常値 (isFraud=0): 透明度を下げて背面に
sns.histplot(data=train[train['isFraud'] == 0], x='Res_Area_log',
             color='skyblue', label='Normal (isFraud=0)',
             kde=True, stat="density", element="step", alpha=0.3)

# 不正値 (isFraud=1): 透明度を上げて形を強調
sns.histplot(data=train[train['isFraud'] == 1], x='Res_Area_log',
             color='orange', label='Fraud (isFraud=1)',
             kde=True, stat="density", element="step", alpha=0.6)

# 密集地帯 (12〜17付近) にズームして見やすくする
plt.xlim(12, 17)

plt.title('Zoomed Log-Transformed Resolution Area: Normal vs Fraud')
plt.xlabel('log(Width * Height + 1)')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

print(f"✅ 解像度の面積変換(Res_Area_log)とズーム表示が完了しました。")

# --- test 側にも適用する ---

# 1. 面積変換（さきほどの関数 get_resolution_area を使用）
test['Res_Area'] = test['id_33'].apply(get_resolution_area)

# 2. 対数変換
test['Res_Area_log'] = np.log1p(test['Res_Area'])

print(f"✅ Test側にも Res_Area / Res_Area_log を作成しました。")
print(f"Testの形状: {test.shape}")
```

id_33について

　グラフでは正常値ではまばらに点在している。これは本来であれば縦横ピクセル数は種別で連続していないため当然ではあるが、不正値では比較的連続している。  
　また、高解像度よりは低解像度のほうがやや不正率が高い傾向にある。  
　縦横比を対数変換したものを特徴量と追加することでこの差異をモデルに学習させることで精度向上を図る。

```python
#2-3　特徴量追加  id_33　解像度変換及び対数変換

# 1. 変換関数の定義
def get_resolution_area(res_str):
    if pd.isna(res_str) or res_str == 'nan':
        return np.nan
    try:
        w, h = map(int, str(res_str).split('x'))
        return w * h
    except:
        return np.nan

# 2. train と test の両方に適用
for df in [train, test]:
    # 面積を計算
    res_area = df['id_33'].apply(get_resolution_area)
    # 対数変換（+1処理含む）して新カラム名で追加
    df['resolution_log'] = np.log1p(res_area)

# 3. 完了確認
print(f"✅ train['resolution_log'] 追加完了 (shape: {train.shape})")
print(f"✅ test['resolution_log']  追加完了 (shape: {test.shape})")
print(train[['id_33', 'resolution_log']].head(10))
```

DeviceInfoの特徴量追加へのアプローチ  
1700件超を全ての統一規格で特徴量追加は困難であり、精査するには多大な工数が掛かる。
件数別に割合が多いものについて、特徴量の検討を行う。

```python
#2-4　特徴量変換  ユニーク値の確認(DeviceInfo)　件数把握

import pandas as pd

def analyze_device_frequency_fixed(df, top_n=50):
    """
    Device_Info_Nameを左詰め、数値を整列させて表示する
    """
    # 1. 基本集計（実際に存在するデータのみ）
    summary = df.groupby('DeviceInfo', observed=True)['isFraud'].agg(['count', 'mean']).reset_index()

    # 2. 指標の計算
    total_count = len(df)
    summary['Percent'] = (summary['count'] / total_count) * 100
    summary['Fraud_Rate'] = (summary['mean'] * 100).round(2)

    # 3. 件数順にソートして上位を抽出
    summary = summary.sort_values('count', ascending=False).head(top_n)

    # 4. ヘッダーの表示
    print(f"📊 --- DeviceInfo 頻度ランキング (Top {top_n}) ---")
    header = f"{'Device_Info_Name':<40} | {'Count':>10} | {'Percent(%)':>12} | {'Fraud_Rate(%)':>15}"
    print("-" * len(header))
    print(header)
    print("-" * len(header))

    # 5. 各行をフォーマットして表示
    for _, row in summary.iterrows():
        # デバイス名は40文字分左詰め、数値は右詰め
        name = str(row['DeviceInfo'])[:40] # 長すぎる名前はカット
        count = int(row['count'])
        percent = f"{row['Percent']:.4f}"
        f_rate = f"{row['Fraud_Rate']:.2f}"

        print(f"{name:<40} | {count:>10} | {percent:>12} | {f_rate:>15}")

    print("-" * len(header))

    # 📍 欠損値(NaN)の確認
    nan_count = df['DeviceInfo'].isna().sum()
    if nan_count > 0:
        nan_fraud = (df[df['DeviceInfo'].isna()]['isFraud'].mean() * 100).round(2)
        nan_pct = ((nan_count / total_count) * 100).round(2)
        print(f"【📍 欠損値(NaN)】 Count: {nan_count:,} | Percent: {nan_pct}% | Fraud_Rate: {nan_fraud}%")

# 実行
analyze_device_frequency_fixed(train)
```

　特徴量追加について検討を行ったが、DeviceInfoのシェアでは一般的なOSのシェアが高いが、それでも8%程度であり、ユニーク値としてはかなり細かく分かれている。  
　また、不正として一貫性はなく、HUAWEIのプロセッサ名表記についても、他の同様のパターンかつシェアが高いものはなかった。  
　そして高額不正においても、NaNの中に紛れているため、有効な特徴量追加には残念ながら至らなかったが、不正手口の手段については知見が深まった。

```python
#3-1　特徴量変換　数値型(int.froat)のカラム確認　基本情報(TransactionDT)

train['TransactionDT'].astype('float64').describe().to_frame().T
```

　最小値が86400であることから、基準日時から1日後が最初のレコードであると思われる。であれば、最大値より、当該データは183日分の記録となる 。
基準日時から1日後が最初のレコード
*   24時間周期 カラム名「time_24h」
*   週換算　　カラム名「time_week」
週換算し、7日周期
*   １日換算　カラム名「time_day」
時間換算し、24h時間周期
*   月換算　　 カラム名「time_month」
30日区切り、端数の3日分は最後の月に参入


また、time_24h および time_week に対しては、23時と0時、あるいは日曜と月曜の連続性を保持するため、sin/cos変換（Cyclical Encoding）を施し、周期的な特徴を追加した。






```python
#3-1　特徴量変換　数値型(int.froat)のカラム確認　基本情報(TransactionDT)

import numpy as np

# 1. 重複防止のリセット
cols_to_drop = [
    'time_24h', 'time_week', 'time_day', 'time_month',
    'time_24h_sin', 'time_24h_cos', 'time_week_sin', 'time_week_cos'
]

for df in [train, test]:
    existing_cols = [c for c in cols_to_drop if c in df.columns]
    df.drop(columns=existing_cols, inplace=True)

print("🗑️ 旧カラムをリセットしました")

# 基準となるTrainの開始時間を取得
train_min_dt = train['TransactionDT'].min()

# 2. 指定の順番で生成 + sin/cos変換 + 月別の動的処理
for df_name, df in [('train', train), ('test', test)]:
    # --- 1. 24時間周期 (0-23.99) ---
    df['time_24h'] = (df['TransactionDT'] / 3600) % 24
    df['time_24h_sin'] = np.sin(2 * np.pi * df['time_24h'] / 24)
    df['time_24h_cos'] = np.cos(2 * np.pi * df['time_24h'] / 24)

    # --- 2. 7日サイクル (0-6: 曜日相当) ---
    # Trainの開始日を基準(Day 0)にするため、TransactionDTから最小値を引く
    df['time_week'] = ((df['TransactionDT'] - train_min_dt) // 86400) % 7
    df['time_week_sin'] = np.sin(2 * np.pi * df['time_week'] / 7)
    df['time_week_cos'] = np.cos(2 * np.pi * df['time_week'] / 7)

    # --- 3. 日別 (基準点からの経過日数) ---
    df['time_day'] = (df['TransactionDT'] - train_min_dt) / 86400

    # --- 4. 月別 (30日単位) ---
    df['time_month'] = (df['time_day'] // 30).astype(int)

    # --- 5. 月別の端数統合ロジック ---
    if df_name == 'train':
        # Trainは5ヶ月目(実質6ヶ月目)以降をすべて 5 に統合
        df.loc[df['time_month'] > 5, 'time_month'] = 5
    else:
        # Test側の判定：最後の月のデータが10日未満なら、その前の月に吸収
        max_month = df['time_month'].max()
        # 最後の月の「開始からの経過日数」を算出
        last_month_duration = df.loc[df['time_month'] == max_month, 'time_day'].max() - (max_month * 30)

        if last_month_duration < 10:
            df.loc[df['time_month'] == max_month, 'time_month'] = max_month - 1
            print(f"💡 Test最終月(Month {max_month})は {last_month_duration:.1f}日のため、Month {max_month-1}に統合しました")

print("✅ カラムを再定義しました（Test側の端数処理も完了）")
print(f"📊 Trainの月構成: {sorted(train['time_month'].unique())}")
print(f"📊 Testの月構成 : {sorted(test['time_month'].unique())}")
```

```python
#3-1　特徴量変換　数値型(int.froat)のカラム確認　基本情報(TransactionDT)

import matplotlib.pyplot as plt

# グラフの土台作成（4段構成）
fig, axes = plt.subplots(4, 1, figsize=(10, 18))

time_features = ['time_24h', 'time_week', 'time_day', 'time_month']
titles = [
    '1. Hourly Cycle (1-24H)',
    '2. Weekly Cycle (1-7 Days)',
    '3. Daily Timeline (1-183 Days)',
    '4. Monthly Timeline (1-6 Months)'
]

for i, col in enumerate(time_features):
    df_plot = train[[col, 'isFraud']].copy()

    # 全て整数に丸め、かつ「+1」して起点を1にする
    df_plot[col] = df_plot[col].astype(int) + 1

    # 集計：件数と不正率
    stats = df_plot.groupby(col)['isFraud'].agg(['count', 'mean']).reset_index()
    stats['fraud_count'] = stats['count'] * stats['mean']
    stats['normal_count'] = stats['count'] - stats['fraud_count']

    # --- 第1軸：件数（積立棒グラフ） ---
    ax1 = axes[i]
    ax1.bar(stats[col], stats['normal_count'], label='Normal', color='skyblue', alpha=0.6)
    ax1.bar(stats[col], stats['fraud_count'], bottom=stats['normal_count'], label='Fraud', color='red', alpha=0.8)
    ax1.set_ylabel('Transaction Volume')
    ax1.set_title(titles[i], fontsize=14, fontweight='bold')
    ax1.legend(loc='upper left')

    # --- 第2軸：不正率（折れ線グラフ） ---
    ax2 = ax1.twinx()
    ax2.plot(stats[col], stats['mean'], color='darkred', marker='o', markersize=3, linewidth=1.5, label='Fraud Rate')
    ax2.set_ylabel('Fraud Rate (Mean)')
    ax2.legend(loc='upper right')

    # 横軸の範囲を適切に設定
    ax1.set_xlim(stats[col].min() - 0.5, stats[col].max() + 0.5)

    # グリッド設定
    ax1.grid(True, which='both', axis='both', linestyle='--', alpha=0.4)

plt.tight_layout()
```

*   月別・日別グラフ  
初月においては取引金額が最大にも関わらず、不正率がかなり低い、以降は急増し、5か月目から少し減少している。
また、日別でみると上下が激しく、不正を積極的に行う日とそうでない日がある。これは日常的に不正を行うグループと特定の日に行うグループで別れているものと思料される。
